```python
import pandas
from pydriller import Repository
import pandas as pd
import matplotlib.pyplot as plt
from quarry import _count_lines
```

```python
path = "https://github.com/mastodon/mastodon.git"
```

```python
data = {}
#reusing antons commit to dictionary code
counter = 0
for commit in Repository(path).traverse_commits():
    counter = counter + 1
    if(counter%100 == 0):
        print(counter, "/", "12366") #display progress
        #break #for testing
    if(counter > 1000):
        break
    for file in commit.modified_files:
        file_path = file.old_path
        if file_path is None: file_path = file.new_path
        if file_path.startswith("app/services"): #for script use \\
            if data.get(file.filename) is None:
                data[file.filename] = {}
            data[file.filename][commit.hash] = _count_lines(str(file.content))  # nested dictionary
```

```python
data
```

```python
df_dict = pd.DataFrame.from_dict(data).T
df_dict = df_dict.reindex((df_dict.mean(axis=1).sort_values(ascending = False).index), axis=0)
df_dict
```

```python
#for each column(commit) go through all rows and store in list all the changed files (!=NaN)
#for every file in list +1 data frame at column (file1) and row of every other file in the list (file2)

index = 0
list = []
for column in df_dict:
    list.append(df_dict[column].dropna().index.values)

#list = list[:6]
list
```

```python
matrix = pd.DataFrame(data = 0, index = df_dict.index.values, columns = df_dict.index.values)
matrix
```

```python
for commit in list:
    for file_name1 in commit:
        for file_name2 in commit:
            matrix.loc[file_name1][file_name2] += 1

```
